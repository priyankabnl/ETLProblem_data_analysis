```python
databp = pd.read_csv('/kaggle/input/practiceone/BloodPressure.csv')
databp.head()
```

```python
datademo = pd.read_csv('/kaggle/input/practiceone/Demographics.csv')
datademo.head()
```

```python
data = pd.merge(datademo,databp,on=["PERSON_ID"])
data.tail()
```

```python
data['TIME'] = pd.to_datetime(data['TIME'])
```

```python
print(data.shape)
print(data.dtypes)
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['PERSON_ID'].unique()
```

```python
data.dropna(inplace=True)
data.shape
```

```python
data_error = data[data['SYSTOLIC_BLOOD_PRESSURE'] == 'ERROR'].index
```

```python
data.drop(data_error, inplace=True)
```

```python
data.shape
```

```python
data_duplicate = data[data.duplicated()]
print(data_duplicate)
```

```python
data.drop_duplicates(inplace = True)
data.shape
```

```python
data['start'] = pd.DatetimeIndex(data['SERVICE_DATE']).normalize()
data['end']= pd.DatetimeIndex(data['TIME']).normalize()
```

```python
data.head()
```

```python
data['result'] = (data['start'] == data['end'])
```

```python
data.tail()
```

```python
data_err = data[data['result'] == False].index
```

```python
data.drop(data_err, inplace=True)
```

```python
delete_columns = ['start', 'end', 'result']
data.drop(delete_columns,axis =1, inplace = True)
```

```python
data.shape
```

```python
data['start_time']= pd.DatetimeIndex(data['TIME']).hour
print(data['start_time'])
```

```python
data['end_time']= pd.DatetimeIndex(data['TIME']).minute
print(data['end_time'])
```

```python
data.tail()
```

```python
data['SYSTOLIC_BLOOD_PRESSURE'] = pd.to_numeric(data['SYSTOLIC_BLOOD_PRESSURE'])
```

```python
conditions = [
    (data['AGE_MONTHS'] >= 44) & (data['SYSTOLIC_BLOOD_PRESSURE'] <= 55),
    (data['AGE_MONTHS'] < 44) & (data['SYSTOLIC_BLOOD_PRESSURE'] <= 46)
]
```

```python
values = ['tier_1', 'tier_2']
```

```python
data['tier'] = np.select(conditions, values)
```

```python
data_err = data[data['tier'] == '0'].index
data.drop(data_err, inplace=True)
```

```python
data.info()
```

```python
data = data.reset_index(drop=True)
data.head()
```

```python
s = data["end_time"] - np.arange(len(data))
data["consecutive"] = s.groupby(s).cumcount() + 1
```

```python
data.head()
```

```python
with open('output.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(['PERSON_ID', 'SERVICE_DATE', 'DURATION_IN_MINS', 'SYSTOLIC_BLOOD_PRESSURE'])
    for index, row in data.head(data.shape[0] - 14).iterrows():
        if data.loc[index, 'consecutive'] == 1 and data.loc[index + 13, 'consecutive'] == 14:
            for i in range(0,14):
                product = (data.loc[index+i, 'PERSON_ID'],data.loc[index+i, 'SERVICE_DATE'],data.loc[index+i, 'end_time'],data.loc[index+i, 'SYSTOLIC_BLOOD_PRESSURE'])
                writer.writerow(product)            
```
